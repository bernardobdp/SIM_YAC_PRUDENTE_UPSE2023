```python
#SIMULADOR ECUACION DE LA DIFUSIVIDAD
import numpy as np
```

```python
P=[0, 4000, 4000, 4000, 0] #CONDICIONES INICIALES
Nodos=len(P)
Nodos2=[0,1,2,3,4]
Tiempo=1000
DT=10
DX=200

t=DT
Pin=P
Pin2=Pin
Pdt=[0,0,0,0,0]
```

```python
def Presdt(P1,P2,P3,deltat,deltax):
  return(((P1-2*P2+P3)/(deltax*deltax))*deltat)+P2
```

```python
Presdt(P[0],P[1],P[2],DT,DX)
```

```python
while t<Tiempo:
     for i in(Nodos2):
       if i==0:
         Pdt[i]=P[0]
       elif i==len(Nodos2)-1:
         Pdt[i]=P[len(Nodos2)-1]
       else:
         Pdt[i]=Presdt(Pin[i-1],Pin[i],Pin[i+1],DT,DX)
    
     t=t+DT
     Pdt=Pin
     Pin2=np.append(Pin2,Pdt)
```
